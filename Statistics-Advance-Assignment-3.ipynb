Q1: What is Estimation Statistics? Explain point estimate and interval estimate.

Answer(Q1):

Estimation statistics is a branch of inferential statistics that deals with making predictions or inferences about population parameters based on sample data. In many real-world situations, it is often impractical or impossible to collect data from an entire population, so researchers use a sample to estimate the unknown population parameters.

There are two main types of estimation: point estimation and interval estimation.

1. Point Estimate:
A point estimate is a single numerical value that is used to estimate an unknown population parameter. It is derived from the sample data and provides an approximation of the true value of the parameter. The process of obtaining a point estimate involves using sample statistics (such as the sample mean, sample proportion, or sample standard deviation) as estimators for their corresponding population parameters (population mean, population proportion, or population standard deviation).

For example, if we want to estimate the average height of all students in a university, we can collect a sample of students' heights and calculate the sample mean height. This sample mean will serve as our point estimate for the population mean height.

Point estimates are simple and easy to compute but provide limited information about the accuracy of the estimate. They do not give any indication of how close the estimate is to the true value of the parameter.

2. Interval Estimate:
An interval estimate, on the other hand, provides a range of values within which the true population parameter is likely to lie, along with a level of confidence associated with that range. It accounts for the uncertainty in the estimation process and provides more information about the accuracy and reliability of the estimate compared to point estimates.

The most common type of interval estimate is the confidence interval. A confidence interval is a range of values constructed around the point estimate, such that we can be reasonably confident (with a specified level of confidence, usually expressed as a percentage) that the true population parameter lies within this interval. The level of confidence is typically chosen to be 95% or 99%, but it can be adjusted based on the researcher's preference.

For example, if we have a 95% confidence interval for the average height of students in a university, it might be expressed as "The average height of students is between 165 cm and 175 cm with 95% confidence."

The width of the confidence interval depends on the variability of the data and the chosen level of confidence. A wider interval indicates greater uncertainty, while a narrower interval suggests a more precise estimate.

In summary, estimation statistics involves using sample data to make educated guesses about population parameters. Point estimates give single values as approximations, while interval estimates provide a range of values with a level of confidence that the true parameter lies within that range. Interval estimates are generally preferred when reporting estimates, as they give a better sense of the uncertainty associated with the estimation process.

Q3: What is Hypothesis testing? Why is it used? State the importance of Hypothesis testing.

Answer(Q3):

Hypothesis testing is a statistical method used to make inferences about population parameters based on sample data. It involves formulating two competing hypotheses about a population parameter and then using sample data to determine which hypothesis is more likely or supported by the evidence. The two main hypotheses involved in hypothesis testing are the null hypothesis (H0) and the alternative hypothesis (Ha).

The process of hypothesis testing typically involves the following steps:

1. Formulate the Null Hypothesis (H0): The null hypothesis represents the default or status quo assumption about the population parameter. It states that there is no significant difference or effect, or no relationship between variables in the population.

2. Formulate the Alternative Hypothesis (Ha): The alternative hypothesis represents the claim or assertion that contradicts the null hypothesis. It is the hypothesis that researchers aim to provide evidence for, suggesting that there is a significant difference, effect, or relationship in the population.

3. Collect Sample Data: Researchers gather data from a sample, which is a subset of the population of interest.

4. Test Statistic: A test statistic is calculated from the sample data, which serves as a numerical summary to help determine how well the sample data aligns with the null hypothesis.

5. Determine the Rejection Region/Critical Value: Based on the test statistic, researchers define a critical region or critical value that helps decide whether to reject the null hypothesis in favor of the alternative hypothesis.

6. Make a Decision: Compare the test statistic to the critical value or critical region. If the test statistic falls into the critical region, the null hypothesis is rejected in favor of the alternative hypothesis. If it does not fall into the critical region, there is insufficient evidence to reject the null hypothesis, and it is retained.

Importance of Hypothesis Testing:

1. Inference about Population: Hypothesis testing allows researchers to make inferences about population parameters when they only have access to sample data. It helps them draw conclusions and make predictions beyond the data they have collected.

2. Objectivity and Rigor: Hypothesis testing provides a structured and objective approach to evaluate hypotheses. It helps avoid making conclusions based solely on intuition or personal beliefs.

3. Scientific Validity: Hypothesis testing is an essential part of the scientific method. It helps ensure that research findings are reliable and valid by subjecting them to statistical scrutiny.

4. Decision Making: Hypothesis testing aids in making important decisions in various fields, such as medicine, economics, engineering, and social sciences. For example, it can be used to determine the effectiveness of a new drug or the impact of a policy change.

5. Identifying Relationships: Hypothesis testing is used to determine whether there is a significant relationship between variables. This is crucial for understanding cause-and-effect relationships and making evidence-based decisions.

In summary, hypothesis testing is a fundamental tool in statistical analysis that enables researchers to draw conclusions about population parameters from sample data, make informed decisions, and validate scientific findings. It plays a crucial role in ensuring the rigor and validity of research in various disciplines.

Q5. Write a Python script to conduct a hypothesis test on the difference between two population means,
given a sample from each population.

Answer(Q5):



```python
import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2, alpha=0.05):
    """
    Conducts a two-sample t-test on the difference between two population means.

    Parameters:
        sample1 (array-like): The first sample data.
        sample2 (array-like): The second sample data.
        alpha (float, optional): The significance level for the test (default is 0.05).

    Returns:
        result (dict): A dictionary containing the test result.
                       - 't_statistic': The t-statistic value.
                       - 'p_value': The p-value.
                       - 'reject_null': True if the null hypothesis is rejected, False otherwise.
    """
    # Calculate the sample means and standard deviations
    
    mean1 = np.mean(sample1)
    mean2 = np.mean(sample2)
    std1 = np.std(sample1, ddof=1)  # ddof=1 for unbiased estimation
    std2 = np.std(sample2, ddof=1)

    # Calculate the sample sizes
    n1 = len(sample1)
    n2 = len(sample2)

    # Calculate the pooled standard error
    pooled_std = np.sqrt((std1**2 / n1) + (std2**2 / n2))

    # Calculate the t-statistic
    t_statistic = (mean1 - mean2) / pooled_std

    # Calculate the degrees of freedom
    df = n1 + n2 - 2

    # Calculate the p-value (two-tailed test)
    p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), df))

    # Compare p-value with significance level (alpha) to make a decision
    reject_null = p_value < alpha

    result = {
        't_statistic': t_statistic,
        'p_value': p_value,
        'reject_null': reject_null
    }

    return result

# Example usage:
sample1 = [14, 16, 17, 20, 22, 25, 16, 19, 21, 24]
sample2 = [18, 21, 22, 20, 24, 26, 23, 19, 20, 25]

result = two_sample_t_test(sample1, sample2)

print("Results of Two-Sample T-Test:")
print("T-Statistic:", result['t_statistic'])
print("P-Value:", result['p_value'])
print("Reject Null Hypothesis:", result['reject_null'])
```

Q8. Define p-value and explain its significance in hypothesis testing.

Answer(Q8):
    
The p-value, short for "probability value," is a crucial concept in hypothesis testing. It quantifies the strength of evidence against the null hypothesis (H0) based on the observed sample data. The p-value represents the probability of obtaining a test statistic as extreme as, or more extreme than, the one observed, assuming that the null hypothesis is true.

In hypothesis testing, we set up two competing hypotheses:

1. Null Hypothesis (H0): This is the default assumption or the statement of no effect, no difference, or no relationship between variables. It is often denoted as "H0: μ = μ0" for testing a population mean, where μ is the population mean, and μ0 is the hypothesized value.

2. Alternative Hypothesis (Ha): This is the statement of the effect, difference, or relationship that we want to test or show evidence for. It can be a one-tailed (directional) or two-tailed (non-directional) hypothesis.

After collecting and analyzing the sample data, we calculate the test statistic (e.g., t-statistic, z-statistic, F-statistic) and then determine the p-value associated with that test statistic.

The significance of the p-value lies in the following interpretation:

- If the p-value is small (typically smaller than the chosen significance level alpha, usually 0.05), it indicates that the observed data is unlikely to have occurred under the assumption of the null hypothesis being true. In other words, there is strong evidence against the null hypothesis.
- If the p-value is large (greater than or equal to alpha), it suggests that the observed data is reasonably likely to have occurred under the assumption of the null hypothesis being true. In this case, we fail to reject the null hypothesis.

The decision to reject or fail to reject the null hypothesis is based on comparing the p-value with the significance level (alpha). The general rule is:

- If p-value ≤ alpha: Reject the null hypothesis (there is enough evidence to support the alternative hypothesis).
- If p-value > alpha: Fail to reject the null hypothesis (there is not enough evidence to support the alternative hypothesis).

It is important to note that a significant result (small p-value) does not prove the alternative hypothesis is true; it merely suggests that the data provides evidence against the null hypothesis. Additionally, the p-value does not quantify the size or practical significance of the effect; it only indicates the strength of evidence for or against the null hypothesis based on the observed sample data. Therefore, researchers should consider the context and practical implications when interpreting the results of a hypothesis test.

Q10. Write a Python program to calculate the two-sample t-test for independent samples, given two random samples of equal size and a null hypothesis that the population means are equal.


Answer(Q10):


```python
import numpy as np
from scipy import stats

def two_sample_t_test(sample1, sample2):
    """
    Performs a two-sample t-test for independent samples.

    Parameters:
        sample1 (array-like): The first sample data.
        sample2 (array-like): The second sample data.

    Returns:
        result (tuple): A tuple containing the t-statistic and the two-tailed p-value.
    """
    t_statistic, p_value = stats.ttest_ind(sample1, sample2)
    return t_statistic, p_value

# Example usage:
# Replace these samples with we own data
sample1 = [14, 16, 17, 20, 22, 25, 16, 19, 21, 24]
sample2 = [18, 21, 22, 20, 24, 26, 23, 19, 20, 25]

# Ensure that both samples have equal size
if len(sample1) != len(sample2):
    raise ValueError("Both samples must have the same size.")

# Calculate the two-sample t-test
t_statistic, p_value = two_sample_t_test(sample1, sample2)

# Print the results
print("Results of Two-Sample T-Test:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Check if the null hypothesis can be rejected
alpha = 0.05
if p_value <= alpha:
    print("Reject the null hypothesis. There is enough evidence to suggest that the population means are not equal.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest that the population means are different.")
```

Q11: What is Student’s t distribution? When to use the t-Distribution.


Answer(Q11);

Student's t-distribution, also known as the t-distribution, is a probability distribution used in statistical hypothesis testing and interval estimation. It is similar to the standard normal distribution (z-distribution) but is better suited for small sample sizes or when the population standard deviation is unknown.

The t-distribution is characterized by its degrees of freedom (df), which determines the shape of the distribution. The degrees of freedom are related to the sample size used to estimate a population parameter. The more degrees of freedom, the closer the t-distribution resembles the standard normal distribution.

The t-distribution is bell-shaped and symmetric, centered around zero (mean = 0). However, it has heavier tails compared to the standard normal distribution. As the degrees of freedom increase, the t-distribution approaches the standard normal distribution.

When to use the t-distribution:
1. **Small Sample Sizes:** When dealing with small sample sizes (typically n < 30) or when the population standard deviation is unknown, the t-distribution is more appropriate than the standard normal distribution. For small samples, using the z-distribution (standard normal) could lead to inaccurate results.

2. **Estimating Population Parameters:** The t-distribution is used when estimating population parameters (e.g., population mean) based on sample data. In this case, we use the t-distribution to calculate confidence intervals or perform hypothesis tests.

3. **Comparing Means of Independent Samples:** When comparing the means of two independent samples (e.g., from two different groups or populations) with small sample sizes, the two-sample t-test is used. The t-distribution is essential for this type of hypothesis testing.

4. **Regression Analysis:** In linear regression analysis, the t-distribution is used to test the significance of regression coefficients.

In summary, the t-distribution is a valuable statistical tool when dealing with small sample sizes or situations where the population standard deviation is unknown. It accounts for the additional uncertainty arising from estimating the population parameters based on limited sample data. When working with large sample sizes (n ≥ 30) and when the population standard deviation is known, the t-distribution converges to the standard normal distribution, making the use of the z-distribution more appropriate.

Q12: What is t-statistic? State the formula for t-statistic.


Answer(Q12):


The t-statistic (or t-value) is a measure used in hypothesis testing to assess the difference between sample means and the null hypothesis. It indicates how many standard errors the sample mean differs from the hypothesized population mean under the assumption that the null hypothesis is true.

The formula for calculating the t-statistic for a single sample is as follows:

t = (x̄ - μ) / (s / √n)

Where:
- t is the t-statistic.
- x̄ is the sample mean.
- μ is the hypothesized population mean (under the null hypothesis).
- s is the sample standard deviation.
- n is the sample size.

The t-statistic measures the distance between the sample mean (x̄) and the hypothesized population mean (μ) in terms of standard errors (s/√n). If the t-statistic is large, it suggests that the observed difference between the sample mean and the hypothesized mean is unlikely to occur by random chance alone, providing evidence against the null hypothesis.

In the context of a two-sample t-test for independent samples (comparing means of two groups), the formula for the t-statistic is slightly different, but the concept remains the same. It compares the difference between the two sample means to the variability within the samples and provides evidence of whether the two population means are different.

For the two-sample t-test, the formula for calculating the t-statistic is as follows:

t = (x̄1 - x̄2) / √((s1^2 / n1) + (s2^2 / n2))

Where:
- t is the t-statistic for the two-sample t-test.
- x̄1 and x̄2 are the sample means of the two independent samples.
- s1 and s2 are the sample standard deviations of the two independent samples.
- n1 and n2 are the sample sizes of the two independent samples.

The t-statistic is then used to calculate the p-value, which is compared to a significance level (e.g., 0.05) to determine whether the null hypothesis should be rejected or not in hypothesis testing.


Q13. A coffee shop owner wants to estimate the average daily revenue for their shop. They take a random sample of 50 days and find the sample mean revenue to be $500  with a standard deviation of  $50. Estimate the population mean revenue with a 95% confidence interval.


Answer(Q13):

sample_mean = $500 (the sample mean revenue)

t_critical is the critical value from the t-distribution for a 95% confidence level with (n - 1) degrees of freedom

sample_standard_deviation = $50 (the sample standard deviation)

n = 50 (the sample size)

To find the critical value (t_critical) from the t-distribution table or using a statistical software or library, we need to consider a 95% confidence level with 49 degrees of freedom (n - 1 = 50 - 1 = 49).

Using a t-distribution table or statistical software, the critical value for a 95% confidence level and 49 degrees of freedom is approximately 2.009.



```python
import numpy as np

# Given data
sample_mean = 500
sample_standard_deviation = 50
sample_size = 50

# Calculate the critical value from the t-distribution
t_critical = 2.009

# Calculate the standard error (standard deviation of the sample mean)
standard_error = sample_standard_deviation / np.sqrt(sample_size)

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_mean - t_critical * standard_error
upper_bound = sample_mean + t_critical * standard_error

# Display the results
print(f"95% Confidence Interval for the Population Mean Revenue: ${lower_bound:.2f} to ${upper_bound:.2f}")
```

Q14. A researcher hypothesizes that a new drug will decrease blood pressure by 10 mmHg. They conduct a clinical trial with 100 patients and find that the sample mean decrease in blood pressure is 8 mmHg with a standard deviation of 3 mmHg. Test the hypothesis with a significance level of 0.05.


Answer(Q14):

```python
import numpy as np
from scipy import stats

# Given data
sample_mean = 8
sample_standard_deviation = 3
sample_size = 100
hypothesized_mean = 10
alpha = 0.05

# Calculate the t-statistic
t_statistic = (sample_mean - hypothesized_mean) / (sample_standard_deviation / np.sqrt(sample_size))

# Calculate the degrees of freedom (n - 1)
degrees_of_freedom = sample_size - 1

# Calculate the two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), degrees_of_freedom))

# Display the results
print("Results of One-Sample t-test:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare p-value with significance level
if p_value <= alpha:
    print("Reject the null hypothesis. The new drug decreases blood pressure by 10 mmHg.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the new drug decreases blood pressure by 10 mmHg.")
```

Q15. An electronics company produces a certain type of product with a mean weight of 5 pounds and a standard deviation of 0.5 pounds. A random sample of 25 products is taken, and the sample mean weight is found to be 4.8 pounds. Test the hypothesis that the true mean weight of the products is less than 5 pounds with a significance level of 0.01.


Answer(Q15):


```python
import numpy as np
from scipy import stats

# Given data
population_mean = 5
population_standard_deviation = 0.5
sample_size = 25
sample_mean = 4.8
alpha = 0.01

# Calculate the standard error (standard deviation of the sample mean)
standard_error = population_standard_deviation / np.sqrt(sample_size)

# Calculate the t-statistic
t_statistic = (sample_mean - population_mean) / standard_error

# Calculate the degrees of freedom (n - 1)
degrees_of_freedom = sample_size - 1

# Calculate the one-tailed p-value
p_value = stats.t.cdf(t_statistic, degrees_of_freedom)

# Display the results
print("Results of One-Sample t-test (Left-Tailed):")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis. The true mean weight of the products is less than 5 pounds.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the true mean weight of the products is less than 5 pounds.")
```

Q16. Two groups of students are given different study materials to prepare for a test. The first group (n1 = 30) has a mean score of 80 with a standard deviation of 10, and the second group (n2 = 40) has a mean score of 75 with a standard deviation of 8. Test the hypothesis that the population means for the two groups are equal with a significance level of 0.01.


Answer(Q16):

To test the hypothesis that the population means for the two groups are equal, we can perform a two-sample independent t-test. The null hypothesis (H0) is that the two population means are equal (μ1 = μ2), and the alternative hypothesis (Ha) is that the two population means are not equal (μ1 ≠ μ2).

Given:
- Sample size of the first group (n1) = 30
- Sample mean of the first group (sample_mean1) = 80
- Sample standard deviation of the first group (sample_standard_deviation1) = 10
- Sample size of the second group (n2) = 40
- Sample mean of the second group (sample_mean2) = 75
- Sample standard deviation of the second group (sample_standard_deviation2) = 8
- Significance level (alpha) = 0.01

The formula for the two-sample t-test is as follows:

t = (sample_mean1 - sample_mean2) / √((sample_standard_deviation1^2 / n1) + (sample_standard_deviation2^2 / n2))

Let's calculate the t-statistic and the p-value and then compare the p-value with the significance level to make a decision:

Based on the calculation, if the p-value is less than the significance level (alpha), we reject the null hypothesis. In this case, the p-value would be compared to the significance level (0.01) to draw a conclusion about whether the population means for the two groups are equal. If the p-value is greater than or equal to the significance level, we fail to reject the null hypothesis.

```python
import numpy as np
from scipy import stats

# Given data for Group 1
n1 = 30
sample_mean1 = 80
sample_standard_deviation1 = 10

# Given data for Group 2
n2 = 40
sample_mean2 = 75
sample_standard_deviation2 = 8

# Significance level
alpha = 0.01

# Calculate the degrees of freedom
degrees_of_freedom = n1 + n2 - 2

# Calculate the pooled standard error
pooled_std_error = np.sqrt(((sample_standard_deviation1 ** 2) / n1) + ((sample_standard_deviation2 ** 2) / n2))

# Calculate the t-statistic
t_statistic = (sample_mean1 - sample_mean2) / pooled_std_error

# Calculate the two-tailed p-value
p_value = 2 * (1 - stats.t.cdf(np.abs(t_statistic), degrees_of_freedom))

# Display the results
print("Results of Two-Sample Independent t-test:")
print("T-Statistic:", t_statistic)
print("P-Value:", p_value)

# Compare p-value with significance level
if p_value < alpha:
    print("Reject the null hypothesis. The population means for the two groups are not equal.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the population means for the two groups are different.")
    
```

Q17. A marketing company wants to estimate the average number of ads watched by viewers during a TV program. They take a random sample of 50 viewers and find that the sample mean is 4 with a standard deviation of 1.5. Estimate the population mean with a 99% confidence interval.

sample_mean = 4 (the sample mean number of ads watched)

t_critical is the critical value from the t-distribution for a 99% confidence level with (n - 1) degrees of freedom

sample_standard_deviation = 1.5 (the sample standard deviation)

n = 50 (the sample size)

To find the critical value (t_critical) from the t-distribution table or using a statistical software or library, we need to consider a 99% confidence level with 49 degrees of freedom (n - 1 = 50 - 1 = 49).



```python
import numpy as np
from scipy import stats

# Given data
sample_mean = 4
sample_standard_deviation = 1.5
sample_size = 50
confidence_level = 0.99

# Calculate the critical value from the t-distribution
t_critical = stats.t.ppf((1 + confidence_level) / 2, df=sample_size - 1)

# Calculate the standard error (standard deviation of the sample mean)
standard_error = sample_standard_deviation / np.sqrt(sample_size)

# Calculate the lower and upper bounds of the confidence interval
lower_bound = sample_mean - t_critical * standard_error
upper_bound = sample_mean + t_critical * standard_error

# Display the results
print(f"{confidence_level * 100}% Confidence Interval for the Population Mean:")
print(f"{lower_bound:.2f} to {upper_bound:.2f} ads watched")
```
